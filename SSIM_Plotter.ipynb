```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (replace with your actual data)
models = ['MNIST','Fashion MNIST', 'CIFAR10', 'Moon_Images']
test_loss = [0.9642008355982878, 0.8859657738812113,  0.21267712963353214 ,0.7943453976156266]  # Train loss for each model
train_loss = [0.6369954403858268, 0.4694694591484376, 0.08679826161746912,0.15755160579048946]    # Test loss for each model

# Define width for bars
bar_width = 0.35
index = np.arange(len(models))

# Plotting
plt.bar(index, train_loss, bar_width, label='JPEG Compression')
plt.bar(index + bar_width, test_loss, bar_width, label='Autoencoder Based Compression')

# Adding labels and title
plt.xlabel('Datasets')
plt.ylabel('SSIM')
plt.title('SSIM for JPEG and Autoencoder based Compression')
plt.xticks(index + bar_width / 2, models)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
